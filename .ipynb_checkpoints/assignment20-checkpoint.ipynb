```python
# Import required libraries

import pandas as pd
import numpy as np
```

```python
# Load the dataset

df = pd.read_csv("tmdb_5000_movies.csv")

# Display first 5 rows
df.head()
```

```python
# Dataset shape

print("Dataset Shape:", df.shape)

# Column names

print("\nColumn Names:")
print(df.columns)

# First 5 rows

print("\nFirst 5 Rows:")
print(df.head())

# Dataset information

print("\nDataset Information:")
print(df.info())

# Missing values

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Text column used for recommendation

text_column = "overview"

print("Text column selected for recommendation:", text_column)
```

```python
# Import required libraries for text preprocessing

import re
import string
import nltk
from nltk.corpus import stopwords
```

```python
# Download stopwords

nltk.download("stopwords")

stop_words = set(stopwords.words("english"))
```

```python
# Function to clean text

def clean_text(text):
    
    # Handle missing values
    if pd.isnull(text):
        return ""
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove punctuation
    text = text.translate(str.maketrans("", "", string.punctuation))
    
    # Remove numbers
    text = re.sub(r"\d+", "", text)
    
    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    
    # Remove stopwords
    words = text.split()
    words = [word for word in words if word not in stop_words]
    
    return " ".join(words)
```

```python
# Apply preprocessing to overview column

df["clean_overview"] = df["overview"].apply(clean_text)

# Display original and cleaned text

df[["overview", "clean_overview"]].head()
```

```python
# Check missing values after preprocessing

print(df["clean_overview"].isnull().sum())

# Display first 5 cleaned rows

df[["title", "clean_overview"]].head()
```

```python
# ==========================================================
# Task 3: Convert Text into Numerical Features using TF-IDF
# ==========================================================

# Import TF-IDF Vectorizer

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Create TF-IDF Vectorizer

tfidf = TfidfVectorizer(max_features=5000)
```

```python
# Convert cleaned movie descriptions into TF-IDF matrix

tfidf_matrix = tfidf.fit_transform(df["clean_overview"])
```

```python
# Display the shape of TF-IDF matrix

print("TF-IDF Matrix Shape:", tfidf_matrix.shape)
```

```python
# Display first few feature names

feature_names = tfidf.get_feature_names_out()

print(feature_names[:20])
```

```python
# ==========================================================
# Task 4: Calculate Cosine Similarity
# ==========================================================

# Import cosine similarity

from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Calculate cosine similarity matrix

similarity_matrix = cosine_similarity(tfidf_matrix)
```

```python
# Display the shape of similarity matrix

print("Similarity Matrix Shape:", similarity_matrix.shape)
```

```python
# Display similarity scores of the first movie

print(similarity_matrix[0])
```

```python
# Display similarity score between first and second movie

print("Similarity Score:", similarity_matrix[0][1])
```

```python
# ==========================================================
# Task 5: Build Movie Recommendation Function
# ==========================================================

# Function to recommend similar movies

def recommend_movies(movie_name):
    
    # Check if movie exists
    if movie_name not in df["title"].values:
        print("Movie not found in the dataset.")
        return

    # Get movie index
    movie_index = df[df["title"] == movie_name].index[0]

    # Get similarity scores
    similarity_scores = list(enumerate(similarity_matrix[movie_index]))

    # Sort movies based on similarity
    sorted_movies = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    print("Recommended Movies:\n")

    count = 0

    for movie in sorted_movies[1:]:
        index = movie[0]
        print(df.iloc[index]["title"])

        count += 1

        if count == 5:
            break
```

```python
# Test 1

recommend_movies("Avatar")
```

```python
# Test 2

recommend_movies("Batman Begins")
```

```python
# Test 3

recommend_movies("The Dark Knight")
```

```python
# Test with an invalid movie name

recommend_movies("ABC Movie")
```
